```python
#renaming columns
type_of_energy = ['LPG Propane', 'Natural Gas', 'Fuel Oil', 'Kerosene', 'Wood'] #row'lar bunlar olcak
purpose_of_usage = ['Space heating', 'Secondary Space heating','Air Conditioning', 'Water heating', 'Refridgerators', 'Appliances, Electronics, Lighting and Miscellaneous uses']  #row'lar bunlar olcak

kwh_list= ['SPH', 'WTH', 'OTH', 'COL', 'RFG']
space_heating= 'SPH'
water_heating= 'WTH'
other_purposes = 'OTH'
air_conditioning = 'COL'
refrigerator = 'RFG'


kwh  = 'KWH'
btu = 'BTU'
cubic_feet = 'CUFEET'
gallon = 'GALLON'

lpg_propane = 'LP'
natural_gas = 'NG'
fossil = 'FO'
kerosene = 'KER'
electricity = 'EL'


electricity_kwh_list = []
for i in range(0, len(kwh_list)):
    text = str('KWH' + kwh_list[i])
    electricity_kwh_list.append(text)

print(electricity_kwh_list)


```

```python
unit_energy = ['BTU', 'CUFEET', 'GALLON']
type_of_energy=['LP', 'NG', 'FO', 'KER', 'EL']
purpose_list = ['SPH', 'WTH', 'OTH']
energy_var_list = ['DOEID']


unit_energy = ['BTU', 'GALLON']
type_of_energy=['LP', 'NG', 'FO', 'KER', 'EL']
purpose_list = ['SPH', 'WTH', 'OTH']


for energy in unit_energy:
    for type in type_of_energy:
        name= energy+type
        for purpose in purpose_list:
            new_name =name+purpose
            energy_var_list.append(new_name)


energy_var_list.remove('GALLONNGSPH')
energy_var_list.remove('GALLONNGWTH')
energy_var_list.remove('GALLONNGOTH')
energy_var_list.remove('GALLONELSPH')
energy_var_list.remove('GALLONELWTH')
energy_var_list.remove('GALLONELOTH')

attach=['CUFEETNGSPH',
'CUFEETNGWTH',
'CUFEETNGOTH',
'BTUELCOL',
'BTUELRFG',
'BTUWOOD', ]

for item in attach:
    energy_var_list.append(item)


energy_var_list = energy_var_list + electricity_kwh_list
```
